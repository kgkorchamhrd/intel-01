```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
x = np.linspace(-2, 2, 11)
f = lambda x: x ** 2
fx = f(x)
```

```python
print(x)
print(fx)
```

```python
plt.plot(x, fx, '-o')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('This is an example for 1d graph')
plt.show()
```
