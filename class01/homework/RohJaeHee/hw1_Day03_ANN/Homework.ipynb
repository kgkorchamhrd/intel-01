```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-2, 2, 11)
f = lambda x : x ** 2
fx = f(x)
```

```python
print(x)
print(fx)
```

```python
plt.plot(x, fx, '-o')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('This is an example for 1d graph')
plt.show()
```

```python
x = np.linspace(-2, 2, 11)
y = np.linspace(-2, 2, 11)

print(x)
print(y)
```

```python
x, y = np.meshgrid(x,y)
print(x)
print(y)
```

```python
f = lambda x,y : (x-1) ** 2 + (y-1) ** 2
z = f(x,y)
print(z)
```

```python
from mpl_toolkits.mplot3d import Axes3D

ax = plt.axes(projection='3d', elev=50, azim=-50)
ax.plot_surface(x,y,z, cmap=plt.cm.jet)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

plt.show()
```

```python
ax = plt.axes()
ax.contour(x, y, z, levels=np.linspace(0, 20, 20), cmap=plt.cm.jet)
ax.grid()
ax.axis('equal')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
plt.show()
```

```python
import numpy as np
A = np.array([\
             [4,5,2,1],\
             [2,3,8,0],\
             [1,0,7,2]])
x = np.array([1,2,3,4])
b = A@x.reshape(-1, 1)
# b = A@x
print("A@x = ", b)
```
